```python
import pandas as pd
import numpy as np
```

```python
latents_data = pd.read_json('data/latents_results.json')
metadata = pd.read_csv('data/shapenet_metadata.csv')
```

```python
latents_data.head(10)
```

```python
metadata.head(10)
```

```python
metadata['fullId'] = metadata['fullId'].replace(r'3dw.', '', regex=True)
metadata.drop(['up', 'front', 'tags', 'wnsynset'], axis=1, inplace=True)
# metadata.to_csv('data/metadata_trunc.csv', index=False)
metadata.head(10)
```

```python
metadata_trunc = pd.read_csv('data/metadata_trunc.csv')
metadata_trunc.head(10)
metadata_trunc.wnlemmas.value_counts()
```

```python
metadata_combined = metadata_trunc.copy(True)
metadata_combined['wnlemmas'].replace({'pickup' : 'SUV', 
                                       'off-road' : 'SUV', 
                                       'roadster' : 'convertible',
                                       'racing car' : 'sports car',
                                       'muscle car' : 'sports car',
                                       'minivan' : 'SUV',
                                       'hatchback' : 'compact'}, inplace=True)
metadata_combined.wnlemmas.value_counts()
```

```python
latents_data = latents_data.join(metadata_combined.set_index('fullId'), on='names')
```

```python
print(len(latents_data))
latents_data.head(10)
```

```python
latents_data.to_csv('LION_latents.csv', index=False)
```
